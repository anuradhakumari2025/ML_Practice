min_max scaling is also called normalization technique
Here magnitude of data reduce according to min and max values and range lie between 0 & 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:\\vs-code-folder\\ML_Practice\\Bengaluru_House_Data.csv")
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
sns.displot(dataset["price"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#mms = min_max scaler
mms= MinMaxScaler()
mms.fit(dataset[["price"]])
```

```python
mms.transform(dataset[["price"]])
```

```python
dataset["price_min"] = mms.transform(dataset[["price"]])
```

```python
dataset.head(5)
```

```python
plt.figure(figsize=(20,9))#sets the figure size.
plt.subplot(1,2,1)#creates the first subplot
plt.title("Before")
#sns.histplot is used instead of sns.displot. This function works within the current axes created by plt.subplot.
sns.histplot(dataset["price"])
# sns.displot(dataset["price"])

plt.subplot(1,2,2)# creates the second subplot.
plt.title("After")
sns.histplot(dataset["price_min"])
# sns.displot(dataset["price_min"])

plt.show()# displays the plots.
```
