```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:\\vs-code-folder\\ML_Practice\\Bengaluru_House_Data.csv")
```

```python
dataset.describe()
```

```python
sns.boxplot(dataset["price"])
plt.show()
```

```python
plt.figure(figsize=(15,8))#to clearly show the graph
sns.boxplot(x="price",data=dataset)
plt.show()
```

```python
q1 = dataset["price"].quantile(0.25)
q1
```

```python
q3 = dataset["price"].quantile(0.75)
q3
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range,max_range
```

```python
new_dataset = dataset[dataset["price"]<=max_range]
```

```python
dataset.shape
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="price",data=new_dataset)
plt.show()
```
