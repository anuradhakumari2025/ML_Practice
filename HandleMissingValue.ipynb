Filling mode in review_title which is categorical data

```python
dataset["review_title"].mode()
```

```python
dataset["review_title"].fillna(dataset["review_title"].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
#before filling mode in dimension3 there was 105 missing values
```

The above steps will fill values of one column
Using for loop we can fill categorical data in columns which are object

```python
for i in dataset.select_dtypes(include="object").columns:
  print(i)
```

```python
for i in dataset.select_dtypes(include="object").columns:
  dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```
