```python
import pandas as pd
```

Ordinal Encoding can be done through two perfumes:-
a) Sklearn
b) Map Function

Making our own dataset

```python
df = pd.DataFrame({"size" : ["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head(4)
```

```python
#double bracket is used because it is a 2D data
ord_data = [["s","m","l","xl"]]  #making ordered data as we require
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#categories can be other also that is inbuilt in ordinalEncoder categories
oe = OrdinalEncoder(categories=ord_data)  #oe = ordinal encoding
oe.fit(df[["size"]])
```

```python
oe.transform(df[["size"]])
```

```python
df[["en_size"]] = oe.transform(df[["size"]])
```

```python
df
```

Now performing the same function using map function

```python
#while mapping any no can be assigned
# ord_data1 = {"s":0,"m":1,"l" : 2,"xl":3}
ord_data1 = {"s":5,"m":10,"l" :20,"xl":35}
```

```python
df["size"].map(ord_data1)
```

```python
df["en_size_map"] = df["size"].map(ord_data1)
```

```python
df
```

Now we will perform ordinal encoding in big dataset

```python
dataset = pd.read_csv("C:\\vs-code-folder\\ML_Practice\\Bengaluru_House_Data.csv")
```

```python
dataset.head(5)
```

```python
dataset["area_type"].unique()
```

```python
en_area_type = [['Super built-up  Area', 'Plot  Area', 'Built-up  Area',
       'Carpet  Area']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oen = OrdinalEncoder(categories=en_area_type)  #oen = ordinal encoding
oen.fit_transform(dataset[["area_type"]])
```

```python
dataset["en_area_type"] = oen.fit_transform(dataset[["area_type"]])
```

```python
en_area_type = {'Super built-up  Area':1, 'Plot  Area':2, 'Built-up  Area':3,
       'Carpet  Area':4}
```

```python
dataset["area_type"].map(en_area_type)
```

```python
dataset["area_type"]=dataset["area_type"].map(en_area_type)
```
